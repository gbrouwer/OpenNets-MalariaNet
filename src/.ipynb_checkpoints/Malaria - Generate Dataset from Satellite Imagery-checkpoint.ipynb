### Read Data

```python
df = pd.read_csv('../data/malaria.CSV',delimiter=',')
X = df[['latitude','longitude']].values
X = np.array(X)
W = np.zeros((180,360))

for i in range(X.shape[0]):
    x = X[i,0]
    y = X[i,1]
    if (np.isnan(x) == False):
        x = int(x+90)
        y = int(y+180)
        W[x,y] = W[x,y] + 1
W = np.log(W+1)
```
